```python
# Loads JSON Output from Query from file path 'home/Mercs/':
merc_df = pd.read_json('/content/drive/MyDrive/Mercs/mercdata.json')
merc_df.head()
```

```python
# Resolves Structure for 'player' column (dict -> columns):
test_df_2 = pd.json_normalize(merc_df['player'])
#test_df_2.head()

# Resolves Structure for 'data' column (dict -> columns):
test_df_3 = pd.json_normalize(merc_df['data'])
#test_df_3.head()

# Concatenates newly created 'player' and 'data' dfs:
test_df_concat = pd.concat([test_df_2, test_df_3], axis=1)
#test_df_concat.head()

# Drops extra columns:
test_df_concat.drop(columns=['id', 'username', 'build', 'status', 'country', 'patron', 'ttm', 'tt200m', 'registeredAt', 'updatedAt', 'lastChangedAt', 'lastImportedAt', 'rank', 'type'], inplace=True)

# Writes to CSV:
test_df_concat.to_csv('/content/drive/MyDrive/Mercs/mercdataformatted.csv')
```
